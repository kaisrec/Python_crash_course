```python
gen = range(10)
list_from_gen = list(gen)
print(list_from_gen)  # Output: [0, 1, 2,
list_from_gen.append("kais")
print(list_from_gen)  # Output: [0, 1, 2, ..., 9, 'kais']
a = list_from_gen.pop()
print(f'new list {list_from_gen} popped value: {a}')  # Output: new list[0, 1, 2, ..., 9] popped value: kais
```

```python
a = [7,2,5,1,3 ]
print(sorted(a))  # Output: [1, 2, 3, 5, 7] use function sorted
a.sort()
print(a)  # Output: [1, 2, 3, 5,
```

```python
b = ['saw', 'small', 'He', 'foxes', 'six']
b.sort(key=len)
print(b)  # Output: ['He', 'saw', 'six', 'small', 'foxes']
```
