#### Decimal places

```python
value = 3.14159
print(f"{value:.2f}")    # 3.14
```

#### Thousands separators

```python
num = 1234567
print(f"{num:,}")        # 1,234,567
```

#### Dates and time formatting

```python
from datetime import datetime
now = datetime.now()
print(f"{now:%Y-%m-%d %H:%M}")
```

#### Padding with custom characters

```python
print(f"{42:0>6}")   # 000042 add leading zeros
```

#### Left, center, right alignment

```python
word = "BI"
print(f"{word:<10}")   # left align
print(f"{word:^10}")   # center
print(f"{word:>10}")   # right align
```

##### Debugging Trick

```python
x = 42
print(f"{x=}")        # x=42
```

#### Joining a list

```python
items = [1, 2, 3]
print(f"Items: {', '.join(str(i) for i in items)}")
```

##### Multiline f-strings

```python
name = "Kais"
role = "BI Developer"

msg = f"""
Name: {name}
Role: {role}
"""
print(msg)
```

##### Calling Functions Inside f-strings

```python
def greet(n): return f"Hello, {n}!"

print(f"{greet('Kais')}")
```
