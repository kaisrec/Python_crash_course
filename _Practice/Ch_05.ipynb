```python
items = list(range(0, 40, 3))
for item in items:
    if item % 2 == 0:
        print(f"{item} is even")
    elif item % 1 == 0:
        print(f"{item} is odd")
    elif item % 5 == 0:
        print(f"{item} is multiple of 5")
```

```python
available = ['mushrooms', 'olives', 'green peppers',
                       'pepperoni', 'pineapple', 'extra cheese']

requested = ['mushrooms', 'french fries', 'extra cheese']
for request in requested:
    if request in available:
        print(f"'{request.upper()}' is available")
    else:
        print(f" sorry, '{request.upper()}' is not available")


```
