```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
cars = ['bmw', 'audi', 'toyota', 'subaru']
```

```python
print("Sorted bicycles:", sorted(bicycles))
print("Original bicycles:", bicycles)
print("Sorted cars:", sorted(cars))
print("Original cars:", cars)

bicycles.sort()
cars.sort()  
print("Bicycles after sort():", bicycles)
print("Cars after sort():", cars)
```
